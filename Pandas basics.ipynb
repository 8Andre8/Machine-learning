Проанализируем набор данных по оттоку клиентов телеком-оператора

```python
ds = pd.read_csv("telecom_churn.csv")
ds.head()
```

```python
ds.info()
```

```python
ds.shape
```

Заменим булевый тип 'churn' на целочисленный:

```python
ds['churn'] = ds['churn'].astype('int')
```

Посмотрим, сколько человек уходят, а сколько остаются лояльными:

```python
ds['churn'].value_counts()
```

Видим, что большинство лояльны, но значительная часть оттекает.  
Нормализуем полученные значения:

```python
ds['churn'].value_counts(normalize=True)
```

Таким образом, если бы мы относили всех пользователей к лояльным, то мы бы ошибались в 14,4914% раз, соответсвенно показатель accuracy = 0.855086. Это значение будем считать базовым для анализа (baseline). Дальнейшие исследования должны предлагать решения, обеспечивающие accuracy > 0.855086.  
Сгруппируем данные по лояльности клиентов:

```python
ds.groupby('churn').mean()
```

```python
ds.pivot_table(['total intl calls', 'total intl charge', 'customer service calls'], ['churn'], aggfunc='mean')
```

Уже сейчас видно, что НЕлояльные пользователи чаще звонят в сервисную службу ('customer service calls') и платят больше в роуминге ('total intl charge') за меньшее количество звонков ('total intl calls'). 
  
Рассмотрим как наблюдения в нашей выборке распределены в контексте двух признаков — 'сhurn' и 'international plan'. Для этого построим таблицу сопряженности, воспользовавшись методом *crosstab*:

```python
pd.crosstab(ds['churn'], ds['international plan'])
```

Нормализуем значения:

```python
pd.crosstab(ds['churn'], ds['international plan'], normalize=True)
```

```python
pd.crosstab(ds['churn'], ds['voice mail plan'], normalize=True)
```

Таким образом, видно, что большинство пользователей лояльны и при этом пользуются дополнительными услугами (международного роуминга / голосовой почты), но при этом доля оттока выше у клиентов с подключенным роумингом. Это может означать, что клиенты недовольны неожиданно высокими платами за разговор в роуминге.

```python
pd.crosstab(ds['churn'], ds['international plan'], margins=True)
```

Далее посмотрим на еще один важный признак – "Число обращений в сервисный центр" ('сustomer service calls'). Также построим сводную таблицу:

```python
pd.crosstab(ds['churn'], ds['customer service calls'])
```

Видим, что начиная с 4 звонка, доля оттока клиентов сильно возрастает.  
Добавим еще одну "фичу" булевого типа - много ли раз (>3) клиент звонит в сервис:

```python
ds['many_calls'] = (ds['customer service calls'] >= 4).astype(int)
ds['many_calls']
```

```python
pd.crosstab(ds['many_calls'], ds['churn'])
```

Видим, что доля в этом случае клиент скорее всего уйдет

```python
d = {'yes': True, 'no': False}
ds['international plan'] = ds['international plan'].map(d)
```

```python
pd.crosstab(ds['many_calls'] & ds['international plan'], ds['churn'])
```

```python
pd.crosstab(ds['many_calls'] & ds['international plan'], ds['churn'], normalize=True)
```

Значит, прогнозируя отток клиента в случае, когда число звонков в сервисный центр больше 3 
и подключен роуминг (и прогнозируя лояльность – в противном случае), можно ожидать около 85.8% 
правильных попаданий (accuracy).  
Эти 85.8% > 85.5%, 
которые мы получили с помощью тривиального классификатора (когда считали, что все клиенты лояльны), это новая неплохая отправная точка (baseline) 
для дальнейших моделей машинного обучения.

В dataframe с именем *ds* сохранены данные с 4 колонками: *session_value, group, time, n_users*.  
В переменной *session_value* замените все пропущенные значения на нули.

```python
# заполняем NaN нулями
ds['session_value'].fillna(0) 
# выбираем неотрицательные значения, чтобы по ним рассчитать медиану 
positive = ds.query("n_users >= 0")
# рассчитываем медиану
m = positive.median()
# меняем отрицательные значения на медианные
ds.loc[(ds.n_users < 0), 'n_users'] = m.n_users
```

```python
# еще 1 способ заменить отрицательные значения на медианные 
ds.loc[ds['n_users'] < 0, 'n_users'] = ds.loc[ds['n_users'] >= 0, 'n_users'].median()
```
