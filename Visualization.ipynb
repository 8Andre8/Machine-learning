```python
ds = pd.read_csv('iris.csv', index_col=0)
ds.head()
```

**7. Ещё один важный тип графиков - *pairplot*, отражающий зависимость пар переменных друг от друга, а также распределение каждой из переменных. 
Построим его и разделимна группы с помощью параметра *hue*.**

```python
ax = sns.pairplot(data=ds, hue='species')
```

Из данного графика видно, что параметры *petal length* и *petal wingth* наиболее коррелированы между собой.
