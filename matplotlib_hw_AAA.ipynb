```python
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
```

```python
f, ([ax1, ax2], [ax3, ax4]) = plt.subplots(2, 2)
f.set_figheight(8)  # устанавливаем высоту figure
f.set_figwidth(15)  # устанавливаем ширину figure
f.suptitle('Plot examples', fontsize=20)  # общий заголовок
plt.subplots_adjust(wspace=0.1, hspace=0.25)  # расстояния между графиками

# 1. scatter plot точек из многомерного нормального распределения
mean = [20, 10]
cov = [[1, 0.7], [0.7, 1]]
x1, y1 = np.random.multivariate_normal(mean, cov, 1000).T
ax1.scatter(x1, y1, edgecolors='black')  # рисуем график на первой картинке
ax1.set_title(
    'Scatter plot example')  # устанавливаем заголовок для первой картинки

# 2. гистограмма геометрического распределения
x2 = np.random.geometric(p=0.1, size=3000)
ax2.hist(x2, bins=40, color='r')
ax2.set_title('Hist plot example')

# 3. графики функций log(x) и log(2x)
log_2x = lambda x: np.log(2 * x)
x3 = np.linspace(0.01, 6, 1000)
y3_1 = np.log(x3)
y3_2 = log_2x(x3)
ax3.plot(x3, y3_1, color='k')
ax3.plot(x3, y3_2, color='r')
ax3.set_title('Line plot example')
ax3.legend(['log(x)', 'log(2x)'])  # выводим легенду

# 4. bar plot случайных, нормально распределённых чисел
n = 14  # количество точек для генерации
start_date = dt.datetime(2021, 1, 1).date()
dates = []  
# формируем массив дат для подписей
for x in range(n):
    dates.append(start_date + dt.timedelta(days=x))
x4 = np.random.normal(loc=5, size=n)
ax4.xaxis.set_ticks(dates)
ax4.set_xticklabels(labels=dates, rotation=90)
ax4.bar(dates, x4);
ax4.set_title('Bar plot example');
```
